```python
#Matplotlib

import matplotlib.pyplot as plt
#this line only for Jupyter
%matplotlib inline 
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
```

```python
#Multiple 
plt.subplot(1,3,1)
plt.plot(x,y,'b--')
plt.subplot(1,3,2)
plt.plot(y,x,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g')
plt.show()
```

```python
#Figure object
#Empty Canvas
fig = plt.figure()

#Add set of axes to the figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height

#plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title")

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y,'r')
axes2.set_xlabel("X2")
axes2.set_ylabel("Y2")
axes2.set_title("Title 2")
```

```python
#using figurs and axes in subplots
#similar to plt.figure() but it uses tuple unpacking to grab fig and axes

fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("T")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) 
```

```python
for ax in axes:
    ax.plot(x,y,'b')
fig
```

```python
#To avoid overlapping
for ax in axes:
    ax.plot(x,y,'b')
plt.tight_layout()
fig
```
